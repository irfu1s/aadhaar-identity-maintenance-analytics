combining seprate dataset of enrolment into one dataset using Concatnate

```python
import pandas as pd

df1=pd.read_csv('df1.csv')
df2=pd.read_csv('df2.csv')
df3=pd.read_csv('df3.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
df1.shape, df2.shape, df3.shape
df1.columns.equals(df2.columns)
df2.columns.equals(df3.columns)
```

```python
full_enroll_df = pd.concat([df1, df2, df3], axis=0, ignore_index=True )
```

## Dataset Schemes & Column Meaning
### Aadhaar Enrolment Dataset


| Column | Meaning |
| :--- | :--- |
| `date` | The calendar date on which Aadhaar enrolments were recorded (format: DD-MM-YYYY). Each row represents enrolment activity for that specific day. |
| `state` | The Indian state where the Aadhaar enrolment took place. |
| `district` | The district within the state where the enrolment was recorded. |
| `pincode` | The postal PIN code representing the local enrolment area. |
| `age_0_5` | Number of Aadhaar enrolments for children aged 0 to 5 years in that location on that date. |
| `age_5_17` | Number of Aadhaar enrolments for individuals aged 5 to 17 years in that location on that date. |
| `age_18_greater` | Number of Aadhaar enrolments for individuals aged 18 years and above in that location on that date. |

```python
full_enroll_df.to_csv('full_enrollment_dataset.csv', index=False)
```

## Dataset Schemes & Column Meaning
### Aadhaar Demographic Dataset



This dataset represents aggregated counts of Aadhaar demographic updates across time and geography.
* **Note:** Unlike the Enrolment dataset, this does *not* represent new Aadhaar registrations.
* **Purpose:** It measures how many people updated their demographic details (such as name, date of birth, gender, or mobile number).
* **Granularity:** Each row corresponds to update activity for a specific date, state, district, and PIN code.

| Column | Meaning |
| :--- | :--- |
| `date` | The calendar date on which Aadhaar demographic updates were recorded (format: DD-MM-YYYY). |
| `state` | The Indian state where the update was performed. |
| `district` | The district within the state where the update was recorded. |
| `pincode` | The postal PIN code of the update location. |
| `demo_age_5_17` | Number of Aadhaar demographic updates performed by individuals aged 5 to 17 years in that location on that date. |
| `demo_age_17_plus` | Number of Aadhaar demographic updates performed by individuals aged 17 years and above in that location on that date. |

```python
full_demographic_df.to_csv('Full_demographic_dataset',index=False)
```
